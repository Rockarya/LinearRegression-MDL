```python
import numpy as np
from numpy import array
import matplotlib.pyplot as plt
import pandas as pd
import pickle
from sklearn import linear_model
from sklearn.preprocessing import PolynomialFeatures
split_size = 10
POLY_DEG = 20   # Give the value of degree here upto which you want to calculate

# Read the Train dataset and split it
file = "./data/train.pkl"
fileOpen = open(file, 'rb')
trainData = pickle.load(fileOpen)
arr = np.array(trainData)
new_arr = np.array_split(arr, split_size)

# Read the Test dataset
file = "./data/test.pkl"
fileOpen = open(file, 'rb')
testData = pickle.load(fileOpen)
test_x = testData[:,:-1]
test_y = testData[:,-1]
test_size = len(testData)

VARIANCE = [0]*POLY_DEG
BIAS = [0]*POLY_DEG
IRR_ERROR = [0]*POLY_DEG
MSE = [0]*POLY_DEG
DEGREE = [0]*POLY_DEG

for deg in range(1,POLY_DEG+1):
    # definig array to store predicted values
    predicted_arr = [[0]*test_size] * split_size
    expected_arr = [0]*test_size

    for i in range(len(new_arr)):
        # separate x and y co-ordinates
        x = new_arr[i][:,:-1]
        y = new_arr[i][:,-1]
        polynomial_features = PolynomialFeatures(degree=deg)
        x_poly = polynomial_features.fit_transform(x)


        # do the needful
        reg = linear_model.LinearRegression()
        reg.fit(x_poly, y)
        weight = reg.coef_
        bias = reg.intercept_
        # regression over

        def line(x):
            fun = bias
            temp = 1
            for j in range(deg+1):
                fun += weight[j] * temp
                temp *= x
            return fun


        x_temp = np.linspace(-50, 50, 10000)
        mymodel = list(map(line, x_temp))


    # Remove the comment to print the plots
#         plt.scatter(x, y)
#         plt.plot(x_temp, mymodel, color = 'm')
#         plt.show()
        predicted_arr[i]=list(map(line,test_x))

    # temp1 is for bias and temp2 is for varinace
    temp1=0
    temp2=0
    for i in range(test_size):
        s1=0
        s2=0
        for j in range(split_size):
            s1+=predicted_arr[j][i]
            s2+=(predicted_arr[j][i])**2
        expected_arr[i]=s1/split_size
        temp1+=abs(expected_arr[i] - test_y[i])**2
        temp2+=(s2/split_size) - (expected_arr[i])**2

    BIAS[deg-1] = temp1/test_size
    VARIANCE[deg-1] = temp2/test_size

    squared_error=0   # This squared_error will store all the squared error for MSE and then we will find MSE by dividing by test_size and split_size
    for i in range(split_size):
        for j in range(test_size):
            squared_error+=(predicted_arr[i][j] - test_y[j])**2

    MSE[deg-1] = squared_error/(test_size * split_size)

    IRR_ERROR[deg-1] = MSE[deg-1] - VARIANCE[deg-1] - (BIAS[deg-1])
    DEGREE[deg-1] = deg
    

tabulate_data = {
    "Degree" : DEGREE,
  "Variance": VARIANCE,
  "Bias^2": BIAS,
    "Irreducible Error": IRR_ERROR
}

x = np.linspace(1, 20, 20)
plt.title("Bias-Variance Analysis")
plt.plot(x, BIAS, label="Bias^2")
plt.plot(x, VARIANCE, color = "g", label="Variance")
plt.plot(x, MSE, color = "r", label="MSE")
plt.legend()
plt.xlabel("Degree")
plt.ylabel("Errors")

#load data into a DataFrame object:
df = pd.DataFrame(tabulate_data)
print(df)
```
